```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
x= cancer.data
```

```python
y=cancer.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier()
tree.fit(x_train,y_train)
```

```python
print(f'train score: {tree.score(x_train,y_train)} \ntest score: {tree.score(x_test,y_test)}')
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(max_depth=3)
```

```python
rfc.fit(x_train,y_train)
```

```python
print(f'train score: {rfc.score(x_train,y_train)} \ntest score: {rfc.score(x_test,y_test)}')
```

```python
plt.subplot(2,1,1)
sns.barplot(x=rfc.feature_importances_,y=cancer.feature_names)
plt.subplot(2,1,2)
sns.barplot(x=tree.feature_importances_,y=cancer.feature_names)
```

```python
df2 = pd.DataFrame({'name':cancer.feature_names,'imp':rfc.feature_importances_})
df4 = pd.DataFrame({'name':cancer.feature_names,'imp':tree.feature_importances_})
```

```python
df3 = df2.sort_values('imp',ascending=False)
df5 = df4.sort_values('imp',ascending=False)
```

```python
sns.barplot(data=df3,x='imp',y='name',palette='tab10')
```

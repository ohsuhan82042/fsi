## 시그모이드 함수

```python
import numpy as np
import matplotlib.pyplot as plt

# 시그모이드 활성화 함수와 그 미분 함수 정의
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

```python
# 신경망의 깊이를 설정 (레이어 수)
layer_depths = [1, 2, 4, 8, 16, 32, 64]

# 입력 값
x = np.linspace(-10, 10, 1000)

# 각 레이어에서의 출력 값을 저장할 리스트
outputs = []

# 초기 입력 값을 첫 번째 레이어의 출력으로 설정
layer_output = x

for depth in layer_depths:
    for _ in range(depth):
        layer_output = sigmoid(layer_output)
    outputs.append(layer_output)

# 시각화
plt.figure(figsize=(12, 8))
for i, output in enumerate(outputs):
    plt.plot(x, output, label=f'{layer_depths[i]} layers')

plt.title('Effect of Increasing Layers on Sigmoid Activation Output')
plt.xlabel('Input value')
plt.ylabel('Output value')
plt.legend()
plt.show()
```

```python
# 시그모이드 함수와 그 미분 값 계산
sigmoid_values = sigmoid(x)
sigmoid_derivative_values = sigmoid_derivative(x)

# 시그모이드 함수와 그 미분 값 시각화
plt.figure(figsize=(12, 8))
plt.plot(x, sigmoid_values, label='Sigmoid')
plt.plot(x, sigmoid_derivative_values, label='Sigmoid Derivative')
plt.axvline(0, color='gray', linestyle='--')  # y축에 수직선 추가
plt.title('Sigmoid and Sigmoid Derivative')
plt.xlabel('Input value')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
# 신경망 초기화
np.random.seed(42)
layer_sizes = [1, 10, 10, 1]  # 입력층, 은닉층 2개, 출력층
weights = [np.random.randn(layer_sizes[i], layer_sizes[i+1]) for i in range(len(layer_sizes)-1)]
biases = [np.random.randn(size) for size in layer_sizes[1:]]

# 순전파 함수 정의
def forward_pass(x):
    activations = [x]
    for w, b in zip(weights, biases):
        z = np.dot(activations[-1], w) + b
        a = sigmoid(z)
        activations.append(a)
    return activations

# 역전파 함수 정의
def backward_pass(activations, y_true):
    deltas = [activations[-1] - y_true]
    for i in range(len(weights)-1, 0, -1):
        deltas.append(np.dot(deltas[-1], weights[i].T) * sigmoid_derivative(activations[i]))
    deltas.reverse()
    return deltas

# 초기 입력 값과 목표 값
x_input = np.array([1.0])
y_true = np.array([0.0])

# 순전파 및 역전파
activations = forward_pass(x_input)
deltas = backward_pass(activations, y_true)

# 기울기 값 시각화
plt.figure(figsize=(12, 8))
plt.plot(range(len(deltas)), [np.linalg.norm(delta) for delta in deltas], marker='o')
plt.title('Gradient Magnitude Across Layers')
plt.xlabel('Layer')
plt.ylabel('Gradient Magnitude')
plt.show()
```

## tanh 함수

```python
# tanh 활성화 함수와 그 미분 함수 정의
def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

# 입력 값 생성
x = np.linspace(-10, 10, 1000)

# tanh 함수와 그 미분 값 계산
tanh_values = tanh(x)
tanh_derivative_values = tanh_derivative(x)

# tanh 함수와 그 미분 값 시각화
plt.figure(figsize=(12, 8))
plt.plot(x, tanh_values, label='tanh')
plt.plot(x, tanh_derivative_values, label='tanh Derivative')
plt.axvline(0, color='gray', linestyle='--')  # y축에 수직선 추가
plt.title('tanh and tanh Derivative')
plt.xlabel('Input value')
plt.ylabel('Value')
plt.legend()
plt.show()
```

## ReLU 함수

```python
import numpy as np
import matplotlib.pyplot as plt

# ReLU 활성화 함수와 그 미분 함수 정의
def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

# 입력 값 생성
x = np.linspace(-2, 2, 1000)

# ReLU 함수와 그 미분 값 계산
relu_values = relu(x)
relu_derivative_values = relu_derivative(x)

# ReLU 함수와 그 미분 값 시각화
plt.figure(figsize=(12, 8))
plt.plot(x, relu_values, label='ReLU')
plt.plot(x, relu_derivative_values, label='ReLU Derivative')
plt.axvline(0, color='gray', linestyle='--')  # y축에 수직선 추가
plt.title('ReLU and ReLU Derivative')
plt.xlabel('Input value')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
# 신경망의 깊이를 설정 (레이어 수)
layer_depths = [1, 2, 4, 8, 16, 32, 64]

# 입력 값
x = np.linspace(-10, 10, 1000)

# 각 레이어에서의 출력 값을 저장할 리스트
outputs = []

# 초기 입력 값을 첫 번째 레이어의 출력으로 설정
layer_output = x

for depth in layer_depths:
    for _ in range(depth):
        layer_output = relu(layer_output)
    outputs.append(layer_output)

# 시각화
plt.figure(figsize=(12, 8))
for i, output in enumerate(outputs):
    plt.plot(x, output, label=f'{layer_depths[i]} layers')

plt.title('Effect of Increasing Layers on ReLU Activation Output')
plt.xlabel('Input value')
plt.ylabel('Output value')
plt.legend()
plt.show()
```
